```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
anime = pd.read_csv("anime.csv")
```

```python
anime
```

```python
anime_synopsis = pd.read_csv("anime_with_synopsis.csv")
```

```python
anime_synopsis
```

```python
anime.drop(anime[anime['Genres']=='Hentai'].index)
```

```python
df_full = anime_synopsis.merge(anime, how='inner',on='MAL_ID')
```

```python
df_full
```

```python
df_full.columns
```

```python
df = df_full[['Name_x','Genres_x','sypnopsis','Type']]
```

```python
df = df.rename(columns = {'Genres_x':'Genres','Name_x':'Name','sypnopsis':'synopsis'})
```

```python
df
```

```python
df2 = df.copy()
```

```python
df2['Tags']=df2['Genres']+df2['synopsis']+df2['Type']
```

```python
df2 = df2.drop(columns = ['Genres','synopsis','Type'])
```

```python
df2
```

```python
df2['MAL_ID']=df_full['MAL_ID']
```

```python
df2
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=16214, stop_words='english')
```

```python
vectors = cv.fit_transform(df2['Tags'].values.astype('U')).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
def anime_recommend(name):
    index = df2[df2['Name']==name].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse = True, key = lambda vectors:vectors[1])
    for i in distance[1:6]:
        print(df2.iloc[i[0]].Name)
```

```python
anime_recommend('Trigun')
```
